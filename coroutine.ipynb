asyncio.run() cannot be called when an event loop is already running.

Jupyter/IPython already runs its own event loop in the background, so when you try:

```python
import asyncio

# Coroutine function
async def main():
    print("Start of main coroutine")
    
# main() -> Coroutine object

# Run the main coroutine
asyncio.run(main())
```

## try running on just regular function

```python
import asyncio

# Coroutine function
async def main():
    print("Start of main coroutine")
    
main() #-> Coroutine object 
# This Coroutine object need to be awaited to be executed
```
